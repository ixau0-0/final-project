```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('features_data.csv')
data
```

```python
target = "charges"
y = data[target].values
```

```python
mean = np.sum(y) / len(y)
std = np.sqrt(np.sum((y - mean) ** 2) / len(y))

print("Mean:", mean)
print("Standard:", std)
#المتوسط يعطينا القيمة العامة للتكاليف، والانحراف المعياري يوضح إن فيه تشتت كبير بسبب القيم العالية
```

```python
col = "bmi"
X = data[[col]].values
mean_x = np.mean(X)
std_x = np.std(X)
z_manual = (X - mean_x) / std_x
scaler = StandardScaler()
z_scaler = scaler.fit_transform(X)
compare = pd.DataFrame({
    "manual_z": z_manual.flatten(),
    "scaler_z": z_scaler.flatten()
})
print(compare.head())
#النتيجة اليدوية وقياس الجاهز تقريبًا نفس الشي، وهذا يثبت إن القياس مضبوط
```

```python
num = data.select_dtypes(include=np.number).columns.drop(target)
highest_idx = data[target].idxmax()
lowest_idx = data[target].idxmin()
vec_high = data.loc[highest_idx, num].values.reshape(1, -1)
vec_low = data.loc[lowest_idx, num].values.reshape(1, -1)
cos_sim = cosine_similarity(vec_high, vec_low)[0][0]

print("highest target index:", highest_idx)
print("lowest target index:", lowest_idx)
print("cosine Similarity:", cos_sim)
#هذا يوضح إذا السجل الأعلى والأقل في التكاليف متشابهين أو لا
```

```python
high_smokers = data[data["smoker"] == "yes"]
prob = (high_smokers["charges"] > 20000).mean()

print("Probability:", prob*100)
#يبين لنا كم نسبة الحالات العالية عند المدخنين، وهذا يدعم إن التدخين يرفع التكاليف
```
